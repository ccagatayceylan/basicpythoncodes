```python
def divide (number):
    return number / 2
```

```python
my_list = [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
for num in my_list:
    print(divide(num))
```

```python
# map
```

```python
map(divide, my_list)
```

```python
list(map(divide, my_list))
```

```python
# diğer yöntem
```

```python
my_new_list = []
for num in my_list:
    my_new_list.append(divide(num))
print(my_new_list)
```

```python
def control_string(string):
    return "Rock" in string
```

```python
control_string("Rock")
```

```python
control_string("rock")
```

```python
my_artist_list = ["Rock", "Metal", "Pop", "Rock group"]
```

```python
list(map(control_string, my_artist_list))
```

```python
#filter
```

```python
list(filter(control_string, my_artist_list))
```
