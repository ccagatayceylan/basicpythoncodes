```python
class DogYears():

    year_factor = 7

    def __init__(self, age=4):
        self.age = age

    def calculation(self):
        return self.age * self.year_factor
```

```python
my_dog = DogYears(3)
```

```python
my_dog.calculation()
```

```python
my_dog = DogYears()
```

```python
my_dog.calculation()
```

```python
class Class1():

    def __init__(self):
        print("Class 1 created")
    
    def method_1(self):
        print("method 1")
    
    def method_2(self):
        print("method 2")
```

```python
my_instance = Class1()
```

```python
my_instance.method_1
```

```python
my_instance.method_1()
```

```python
my_instance.method_2()
```

```python
class Class2(Class1):

    def __init__(self):
        Class1.__init__(self)
        print("Class 2 created")

    def method_3(self):
        print("method 3")

    #override

    def method_1(self):
        print("method 1 override")
```

```python
my_instance2 = Class2()
```

```python
my_instance2.method_1
```

```python
my_instance2.method_1()
```

```python
my_instance2.method_3()
```

```python
my_instance2.method_1()
```

```python
my_instance.method_1()
```

```python
class Apple():

    def __init__(self, name):
        self.name = name

    def information(self):
        return self.name + " 100 calories"
```

```python
class Banana():
    def __init__(self, name):
        self.name = name
    
    def information(self):
        return self.name + " 200 calories"
```

```python
banana = Banana("banana")
```

```python
apple = Apple("apple")
```

```python
banana.information()
```

```python
apple.information()
```

```python
fruit_list = [banana, apple]
```

```python
for fruit in fruit_list:
    print(fruit.information())
```

```python
def get_info(fruit):
    print(fruit.information())
```

```python
get_info(banana)
```

```python
get_info(apple)
```
