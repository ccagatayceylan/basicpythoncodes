```python
sport_list = ["jump", "run", "football"]
```

```python
calories_list = [300, 400, 500]
```

```python
days = ["monday", "tuesday", "wednesday"]
```

```python
zip(sport_list, calories_list, days)
```

```python
new_zip_list_output = zip(sport_list, calories_list, days)
```

```python
new_zip_list_output = list(zip(sport_list, calories_list, days))
```

```python
new_zip_list_output
```

```python
for eleman in new_zip_list_output:
    print(eleman)
```
