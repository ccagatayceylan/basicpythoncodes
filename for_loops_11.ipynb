```python
my_list = [1, 2, 3, 4, 5]
```

```python
for number in my_list:
    print(number)
```

```python
for item in my_list:
    new_number = item * 5
    print(new_number)
```

```python
for number in my_list:
    if number % 2 == 0 :
        print(number)
```

```python
if 2 in my_list:
    print("true")
```

```python
for num in my_list:
    if num == 2:
        print("true")
```
