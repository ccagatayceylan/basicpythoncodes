```python
my_list = [10, 20, 30, 40, 50, 60]
```

```python
for number in my_list:
    print(number)
```

```python
for number in my_list:
    print(number * 8)
```

```python
for new_numbers in my_list:
    print(new_numbers * 2)
```

```python
for num in my_list:
    if num == 30:
        break
    print(num * 3)
```

```python
for num in my_list:
    if num == 30:
        continue
    print(num * 3)
```

```python
for num in my_list:
    pass  # hatayı almamak için pass dedik ne yazacağımız belli değil idi...
```
